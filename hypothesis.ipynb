```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python
df = pd.read_excel(r"C:\Users\anushma\OneDrive\Documents\train.xlsx")
```

```python
tech = df[df["Category"]=="Technology"]["Sales"]
furn = df[df["Category"]=="Furniture"]["Sales"]
```

```python
t_stat, p_value = ttest_ind(tech, furn)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
if p_value < 0.05:
    print("Significant difference: Technology sales are higher")
else:
    print("No significant difference")
```
